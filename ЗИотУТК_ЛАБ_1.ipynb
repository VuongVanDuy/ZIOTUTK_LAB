# 1. исследование в двумерном пространстве.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def harmonic_signal(A, f, t, phi0):
    return A * np.cos(2 * np.pi * f * t + phi0)
```

```python
A = 20
f = 1
t = np.linspace(0, 10, 700)
A_var = 2*t
freq_var =  np.pi*t
phi0 = 0
```

```python
#Гармонический сигнал

signal_1 = harmonic_signal(A=A, f=f, t=t, phi0=phi0)
plt.figure(figsize=(10, 4))
plt.plot(t, signal_1)
plt.xlabel('Время (t)')
plt.ylabel('Амплитуда (y)')
plt.title('Гармонический сигнал')
plt.show()
```

```python
#Гармонические сигналы с изменяющейся частотой

signal_2 = harmonic_signal(A=A, f=freq_var, t=t, phi0=phi0)
plt.figure(figsize=(10, 4))
plt.plot(t, signal_2)
plt.xlabel('Время (t)')
plt.ylabel('Амплитуда (y)')
plt.title('Гармонические сигналы с изменяющейся частотой')
plt.show()
```

```python
# Гармонический сигнал с изменяющейся амплитудой

signal_3 = harmonic_signal(A_var, f=f, t=t, phi0=phi0)
plt.figure(figsize=(10, 4))
plt.plot(t, signal_3)
plt.xlabel('Время (t)')
plt.ylabel('Амплитуда (y)')
plt.title('Гармонический сигнал с изменяющейся амплитудой')
plt.show()
```

```python
# Tạo các tín hiệu với các tần số và biên độ khác nhaus
signal_sum = signal_1 + signal_2 + signal_3
plt.figure(figsize=(10, 4))
plt.plot(t, signal_sum)
plt.xlabel('Время (t)')
plt.ylabel('Амплитуда (y)')
plt.title('Сумма сигналов')
plt.show()
```

# 2. исследование в трехмерном пространстве.

```python
from scipy.fft import fft, fftfreq
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Hàm tạo tín hiệu hình sin trong không gian 3D
def harmonic_signal_3d(A, f, t, phi0):
    # Tín hiệu dọc theo trục Z
    z = A * np.cos(2 * np.pi * f * t + phi0)
    x = t  # Thành phần X
    y = np.full_like(t, A)  # Thành phần Y
    return np.vstack((x, y, z))  # Kết hợp thành ma trận 3D
```

```python
# Hàm xoay tín hiệu quanh trục X
def rotate_signal_3d(signal, angle_deg):
    angle_rad = np.deg2rad(angle_deg)  # Đổi góc từ độ sang radian
    rotation_matrix = np.array([
        [1, 0, 0],
        [0, np.cos(angle_rad), -np.sin(angle_rad)],
        [0, np.sin(angle_rad), np.cos(angle_rad)]
    ])
    return np.dot(rotation_matrix, signal)
```

```python
A = 20
f = 1
t = np.linspace(0, 10, 700)
A_var = 2*t
freq_var =  np.pi*t
phi0 = 0
```

```python
signal_1 = harmonic_signal_3d(A, f, t, phi0)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(signal_1[0], signal_1[1], signal_1[2], label='Гармонический сигнал')
ax.set_xlabel('Ox (Время)')
ax.set_ylabel('Oy (Амплитуда)')
ax.set_zlabel('Oz')
ax.set_title('Гармонический сигнал 3D')
ax.legend()
plt.show()
```

```python
signal_2 = harmonic_signal_3d(A, freq_var, t, phi0)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(signal_2[0], signal_2[1], signal_2[2], label='Гармонические сигналы с изменяющейся частотой')
ax.set_xlabel('Ox (Время)')
ax.set_ylabel('Oy (Амплитуда)')
ax.set_zlabel('Oz')
ax.set_title('Гармонические сигналы с изменяющейся частотой 3D')
ax.legend()
plt.show()
```

```python
signal_2_rotated = rotate_signal_3d(signal_2, 4)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(signal_2_rotated[0], signal_2_rotated[1], signal_2_rotated[2], label='Гармонические сигналы с изменяющейся частотой')
ax.set_xlabel('Ox (Время)')
ax.set_ylabel('Oy (Амплитуда)')
ax.set_zlabel('Oz')
ax.set_title('Гармонические сигналы с изменяющейся частотой 3D')
ax.legend()
plt.show()
```

```python
signal_3 = harmonic_signal_3d(A_var, f, t, phi0)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(signal_3[0], signal_3[1], signal_3[2], label='Гармонический сигнал с изменяющейся амплитудой')
ax.set_xlabel('Ox (Время)')
ax.set_ylabel('Oy (Амплитуда)')
ax.set_zlabel('Oz')
ax.set_title('Гармонический сигнал с изменяющейся амплитудой 3D')
ax.legend()
plt.show()
```

```python
signal_3_rotated = rotate_signal_3d(signal_3, 8)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(signal_3_rotated[0], signal_3_rotated[1], signal_3_rotated[2], label='Гармонический сигнал с изменяющейся амплитудой')
ax.set_xlabel('Ox (Время)')
ax.set_ylabel('Oy (Амплитуда)')
ax.set_zlabel('Oz')
ax.set_title('Гармонический сигнал с изменяющейся амплитудой 3D')
ax.legend()
plt.show()
```

```python
signal_sum = signal_1 + signal_2_rotated + signal_3_rotated

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(signal_sum[0], signal_sum[1], signal_sum[2], label='Сумма сигналов')
ax.set_xlabel('Ox (Время)')
ax.set_ylabel('Oy (Амплитуда)')
ax.set_zlabel('Oz')
ax.set_title('Сумма сигналов 3D')
ax.legend()
plt.show()
```

```python
z_signal = signal_sum[2]
N = len(z_signal)  # Количество выборок
fs = len(t) / (t[-1] - t[0])
fft_values = fft(z_signal)  # Быстрое преобразование Фурье (FFT)
frequencies = fftfreq(N, 1/fs)  # Вычисление оси частот

# Оставляем только положительные частоты
positive_freqs = frequencies[:N//2]
amplitude_spectrum = np.abs(fft_values[:N//2]) / (N//2)  # Амплитудный спектр

# Определение амплитуды, частоты и фазы
peak_index = np.argmax(amplitude_spectrum)  # Индекс максимальной амплитуды
estimated_amplitude = 2 * amplitude_spectrum[peak_index]  # Коэффициент 2 для корректировки амплитуды
estimated_frequency = positive_freqs[peak_index]  # Оцененная частота
estimated_phase = np.angle(fft_values[peak_index])  # Фаза сигнала

# Вывод результатов
print(f"Частота дискретизации fs: {fs:.2f} Гц")
print(f"Оцененная амплитуда: {estimated_amplitude:.4f}")
print(f"Оцененная частота: {estimated_frequency:.4f} Гц")
print(f"Оцененная фаза: {estimated_phase:.4f} рад")

# Построение амплитудного спектра
plt.figure(figsize=(8, 5))
plt.plot(positive_freqs, amplitude_spectrum, 'b-', label="Амплитудный спектр")
plt.xlabel("Частота (Гц)")
plt.ylabel("Амплитуда")
plt.title("Амплитудный спектр сигнала")
plt.legend()
plt.grid()
plt.show()
```
